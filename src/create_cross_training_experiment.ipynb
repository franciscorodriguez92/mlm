```python
import pandas as pd
#%%
file_exist_2021 = "../data/input/EXIST2021_dataset-test/EXIST2021_dataset/training/EXIST2022_training.tsv"
df_file_exist_2021= pd.read_table(file_exist_2021, sep="\t")

```

```python
df_file_exist_2021.dtypes
```

```python
file_exist_2022 = "../data/input/EXIST 2022 Dataset/test/test_EXIST2022_labeled.tsv"
df_file_exist_2022= pd.read_table(file_exist_2022, sep="\t")
df_file_exist_2022
```

```python
df_file_exist_2022= df_file_exist_2022.rename(columns={"text_case": "test_case"})
df_file_exist_2022.drop('status_id', axis=1, inplace=True)

```

```python
df_file_exist_2022
```

```python
df_file_exist_all = pd.concat([df_file_exist_2021, df_file_exist_2022])
df_file_exist_all = df_file_exist_all[
    df_file_exist_all['task2']!="desacuerdo"]
```

```python
df_file_exist_all
```

```python
df_file_exist_all['task2'].value_counts()
```

```python
df_file_exist_all_samples = df_file_exist_all.groupby('task2').apply(
    lambda x: x.sample(n=800, random_state=123)).reset_index(drop = True)
```

```python
df_file_exist_all_samples['task2'].value_counts()
```

```python
df_file_exist_all_samples
```

```python
from sklearn.model_selection import train_test_split

train,test = train_test_split(df_file_exist_all_samples, test_size=0.2, random_state=19, stratify=df_file_exist_all_samples['task2'])
```

```python
train['task2'].value_counts()
```

```python
test['task2'].value_counts()
```

```python
train.to_csv("../data/input/crossed_training_train.tsv", 
                                 sep="\t", index=False)
```

```python
test.to_csv("../data/input/crossed_training_test.tsv", 
                                 sep="\t", index=False)
```
